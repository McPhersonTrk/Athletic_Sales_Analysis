### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
'''#A groupby or pivot_table function has been used to create a multi-index DataFrame with the "region", "state", and "city" columns. (10 points)
#The aggregated column has been renamed to reflect the aggregation of the data in the column. (1 point)
#The results are sorted in ascending order to show the top five regions, including the state and city that sold the most products. (4 points)
'''
```

```python
combo_df.columns
```

```python
# Show the number products sold for region, state, and city. (groupby, sum, resetindex)(sort by highest sales)
group_df = combo_df.groupby(['region', 'state', 'city']).sum('units_sold').reset_index()

# Rename the sum to "Total_Products_Sold".
group_df.rename(columns={"units_sold": "Total_Products_Sold"}, inplace=True)

# (sort&) Show the top 5 results.
sort_combo_df= combo_df.sort_values(by='units_sold', ascending=False)
group_df.head(5)
```

#### Using `pivot_table`

```python
combo_df.columns
```

```python
# Show the number products sold for region, state, and city.
rows = ['region', 'state', 'city']
columns = None
values = 'units_sold'
pivot_df = combo_df.pivot_table(values=values, index=rows, columns=columns, aggfunc='sum')
pivot_df.reset_index(inplace=True)
# Rename the "units_sold" column to "Total_Products_Sold"
pivot_df.rename(columns={"units_sold": "Total_Products_Sold"}, inplace=True)

# Show the top 5 results.
sort_pivot_df= pivot_df.sort_values(by='Total_Products_Sold', ascending=False)
sort_pivot_df.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
'''A pivot table is created that has the "invoice_date" column as the index and the "total_sales" column assigned to the values parameter. (10 points)

The aggregated column has been renamed to reflect the aggregation of the data in the column. (1 point)

The resample function is used on the pivot table, the data is placed into daily bins, and the total sales for each day is calculated. (2 points)

The results are sorted in ascending order to show the days that generated the most women's athletic footwear sales. (2 points)'''
```

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
rows = ['invoice_date']
columns = None
values = 'total_sales'
invd_df= combo_df.pivot_table(values=values, index=rows, columns=columns)
invd_df.sort_values(by='total_sales', ascending=False)
# Resample the pivot table into daily bins, and get the total sales for each day.
invd_df.resample('D').sum()
# Optional: Rename the "total_sales" column to "Total Sales"
invd_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the table.
invd_df.head(500).round()
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
'''The resample function is used on the pivot table, the data is placed into weekly bins, and the total sales for each week is calculated. (3 points)

The results are sorted in ascending order to show the weeks that generated the most women's athletic footwear sales. (2 points)'''
```

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
invd_df.resample('W').sum().round()
# Sort the resampled pivot table in ascending order on "Total Sales".
invd_df.sort_values(by='Total Sales', ascending=True)
```
